```python
import json
import pandas as pd
```

```python
filename = 'uzair_tweets.jsonl'
```

```python
data = []
with open(filename,encoding='utf-8') as f:
    for line in f:
        d = json.loads(line)
        good_columns = [d['created_at'], d['id'], d['lang'], d['full_text'], d['user']['verified'], d['user']['id'], 
                        d['user']['location'], d['entities']['hashtags'], d['geo']]
        data.append(good_columns)
```

```python
good_column = ['created_at', 'id', 'lang', 'full_text', 'verified', 'author_id', 'user_loc', 'hashtags', 'tweet_loc']
df = pd.DataFrame(data, columns=good_column)
```

```python
df.to_csv('march_tweets.csv', sep=',', index=False)
```
